# **United States - Crime Rates - 1960 - 2014**

Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy
```

Step 2. Import the dataset from this address.

```python
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv'
```

Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv(csv_url)
crime.head()
```

Step 4. What is the type of the columns?

```python
crime.info()
```

Step 5. Convert the type of the column Year to datetime64


```python
crime['Year'] = pd.to_datetime(crime['Year'],  format = '%Y')
crime.head()
```

Step 6. Set the Year column as the index of the dataframe

```python
crime.set_index('Year', drop=True, inplace=True)
crime.head()
```

Step 7. Delete the Total column

```python
crime.columns
```

```python
crime.drop(columns='Total', inplace=True)
crime.head()
```

Step 8. Group the year by decades and sum the values

```python
crime.groupby((crime.index.year//10)*10).sum()
```

Step 9. What is the most dangerous decade to live in the US?

```python
crime.idxmax(axis=0)
```
