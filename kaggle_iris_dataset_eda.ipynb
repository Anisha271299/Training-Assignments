```python
import pandas as pd
import warnings 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
iris = pd.read_csv("/content/Iris.csv")
```

1.HEAT MAP

```python
plt.figure(figsize=(10,8))
sns.heatmap(iris.corr(),annot=True,cbar=False)
plt.tight_layout(pad=0.5)
plt.show()
```

2.SCATTER PLOT

```python
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

3.RUG PLOT

```python
sns.rugplot(iris['SepalLengthCm'],color='green')
```

4.KDE PLOT

```python
sns.kdeplot(iris['SepalLengthCm'])
```

5.BAR PLOT

```python
sns.barplot(x='SepalLengthCm',y='SepalWidthCm',data=iris,color='yellow')
```

6.STRIP PLOT

```python
sns.stripplot(x='PetalLengthCm',y='PetalWidthCm',data=iris,palette='GnBu')
```

7.DIST PLOT

```python
sns.displot(data=iris, x="SepalWidthCm", hue="SepalLengthCm", kind="kde", fill=True)
```

8.JOINT PLOT

```python
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

9.BOX PLOT

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

10.VIOLIN PLOT

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
```

12.PAIR PLOT

```python
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```
